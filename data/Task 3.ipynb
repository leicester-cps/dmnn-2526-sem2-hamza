```python
customer_spend = df.groupby("Customer ID")["Revenue"].sum()
```

```python
customer_spend
```

```python
invoice_value = df.groupby(["Customer ID","Invoice"])["Revenue"].sum()
avg_invoice_value = invoice_value.groupby("Customer ID").mean()
avg_invoice_value
```

```python
basket_qty= df.groupby(["Customer ID","Invoice"])["Quantity"].sum()
avg_basket_qty = basket_qty.groupby("Customer ID").mean()
avg_basket_qty
```

```python
customer_country = df.groupby("Customer ID")["Country"] \
.agg(lambda x: x.mode()[0])
```

```python
features = pd.DataFrame({
    "customer_spend": customer_spend,
    "avg_order_value": avg_invoice_value,
    "avg_basket_qty": avg_basket_qty,
    "Customer_country": customer_country
})

features.head()
```

```python
num_orders = df.groupby("Customer ID")["Invoice"].nunique()
features["repeat_customer"] = (num_orders > 1).astype(int)
```
